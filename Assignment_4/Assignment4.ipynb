# DataSet Load

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('./data/HouseTS.csv')
df.info()
df.head()
```

# Missing Values

```python
# Count missing
missing = df.isna().sum().sort_values(ascending=False)
print(missing.head(10))

# Visualize
plt.figure(figsize=(12,6))
sns.heatmap(df.isna(), cbar=False, yticklabels=False)
plt.title('Missing Value Pattern')
plt.show()
```

# Data Types & Date Conversion

```python
df['year'] = pd.to_datetime(df['year'])
df.dtypes
```

# Summary Statistics

```python
df.describe().T
```

# Distribution Plots (Histograms + KDE)

```python
num_cols = ['price', 'Per Capita Income', 'Total Housing Units']

for col in num_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col].dropna(), kde=True, bins=50)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
# Outlier Detection Using Boxplots
num_cols = ['price', 'Per Capita Income', 'Total Housing Units']

for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col].dropna())
    plt.title(f'Boxplot of {col}')
    plt.grid(True)
    plt.show()
```

# Correlation Analysis

```python
# Compute correlation matrix
corr = df.corr(numeric_only=True)

# Heatmap of all correlations
plt.figure(figsize=(16, 12))
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Top correlated features with price
target = 'price'
top_corr = corr[target].drop(target).sort_values(ascending=False)

print("Top features positively correlated with price:")
print(top_corr.head(5))

print("\nTop features negatively correlated with price:")
print(top_corr.tail(5))
```

# Time Series Trend

```python
df['year'] = pd.to_datetime(df['year'])  # if not already done

monthly_avg = df.groupby('year')['price'].mean()

plt.figure(figsize=(14, 5))
monthly_avg.plot()
plt.title("Average House Price Over Time")
plt.xlabel("Year-Month")
plt.ylabel("Avg Price")
plt.grid(True)
plt.show()
```

# Scatter Plot to Explore Relationship

```python
sns.scatterplot(data=df, x='Per Capita Income', y='price', alpha=0.3)
plt.title('Price vs Per Capita Income')
plt.grid(True)
plt.show()

sns.scatterplot(data=df, x='Total Housing Units', y='price', alpha=0.3)
plt.title('Price vs Total Housing Units')
plt.grid(True)
plt.show()
```

# Multi-Feture RelationShip

```python
sns.pairplot(df[['price', 'Per Capita Income', 'Total Housing Units']].dropna())
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

# Rolling Average or Price Difference

```python
df = df.sort_values(['zipcode', 'year'])
df['price_change'] = df.groupby('zipcode')['price'].diff()
df['rolling_avg_price'] = df.groupby('zipcode')['price'].transform(lambda x: x.rolling(3, min_periods=1).mean())

sample = df[df['zipcode'] == df['zipcode'].iloc[0]]

plt.figure(figsize=(10, 5))
plt.plot(sample['year'], sample['price'], label='Original')
plt.plot(sample['year'], sample['rolling_avg_price'], label='3-month Rolling Avg')
plt.legend()
plt.title(f"Price Trend for ZIP {sample['zipcode'].iloc[0]}")
plt.grid(True)
plt.show()
```

# Conclusion

This EDA provided valuable insights into the HouseTS dataset:

- Identified missing values and visualized them using a heatmap.
- Explored distributions and detected outliers via histograms and boxplots.
- Found strong correlations (e.g., between `price` and `Per Capita Income`).
- Observed rising trends in house prices over time.
- Used scatter plots and pairplots to uncover relationships between variables.
- Applied rolling averages and price change analysis for temporal trends.
