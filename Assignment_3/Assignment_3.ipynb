# Loading The Data Working on "Titaninc-Dataset.csv"

```python
import seaborn as sns
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

data = pd.read_csv('data/Titanic-Dataset.csv')
data.head()
```

# Data Summary And Missing Values

```python
# DataSet OverView
print("Shape of the DataSet: ", data.shape)
print("\nColumns in the DataSet: \n", data.columns)

# Cheack for Missing Values
print("\nMissing values in each column:\n", data.isnull().sum())

# DataType and Summary
print("\nData types:\n", data.dtypes)
print("\nSummary Statistics:\n", data.describe(include='all'))
```

```python
# Check for missing values
print("\nMissing values in each column:\n", data.isnull().sum())

print("\nPercentage of missing values in each column:\n", (data.isnull().sum()/len('data/Titanic-Dataset.csv'))*100)
```

```python
# create the heatMap
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), 
            cbar=False, 
            cmap='plasma', 
            yticklabels=False)

plt.title('Missing Values Heatmap', fontsize=14)
plt.xlabel('Columns', fontsize=12)
plt.ylabel('Rows (Records)', fontsize=12)
plt.show()
```

# Impute missing values

```python
# Import necessary libraries

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

data_filled = data.copy()

numerical_cols = data_filled.select_dtypes(include=['number']).columns
categorical_cols = data_filled.select_dtypes(exclude=['number']).columns

imputer_num = IterativeImputer(estimator=RandomForestRegressor(), random_state=0)
data_filled[numerical_cols] = imputer_num.fit_transform(data_filled[numerical_cols])

for col in categorical_cols:
    data_filled[col] = data_filled[col].fillna(data_filled[col].mode()[0])

# Check for any remaining missing values
print("\nMissing values after imputation:\n")
print(data_filled.isnull().sum())
```

# Univariate Analysis

```python
# Analyze 'survived'
print("\nSurvival Count:\n", data_filled['Survived'].value_counts())

sns.countplot(x='Survived', data=data_filled, palette='pastel')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

# Analyze 'pclass'
print("\nPclass Distribution:\n", data_filled['Pclass'].value_counts())

sns.countplot(x='Pclass', data=data_filled)
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()

# Analyze 'age'
print("\nAge Statistics:\n", data_filled['Age'].describe())

sns.histplot(x='Age', data=data_filled, kde=True, bins=30, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Analyze 'sex'
print("\nSex Distribution:\n", data_filled['Sex'].value_counts())

sns.countplot(x='Sex', data=data_filled)
plt.title('Sex Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

# Bivariate Analysis

```python
# Analyze survival based on gender
print("\nSurvival by Gender:\n", data_filled.groupby('Sex')['Survived'].value_counts())

sns.countplot(x='Sex', hue='Survived', data=data_filled, palette='Set2')
plt.title('Survival by Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

# Analyze survival based on passenger class
print("\nSurvival by Passenger Class:\n", data_filled.groupby('Pclass')['Survived'].value_counts())

sns.countplot(x='Pclass', hue='Survived', data=data_filled, palette='pastel')
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()

# Analyze survival based on age groups
data_filled['Age_Group'] = pd.cut(data_filled['Age'], bins=[0, 18, 65, 100], labels=['Child', 'Adult', 'Senior'])

print("\nSurvival by Age Group:\n", data_filled.groupby('Age_Group')['Survived'].value_counts())

sns.countplot(x='Age_Group', hue='Survived', data=data_filled, palette='coolwarm')
plt.title('Survival by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()

# Analyze survival based on gender and class
print("\nSurvival by Gender and Class:\n", data_filled.groupby(['Sex', 'Pclass'])['Survived'].value_counts())

sns.catplot(x='Sex', hue='Survived', col='Pclass', kind='count', data=data_filled, palette='Set1')
plt.show()
```

# Survival Analysis

```python
# Survival Based on Embarkation Point
plt.figure(figsize=(10, 5))
sns.countplot(x='Embarked', hue='Survived', data=data_filled, palette='Spectral')
plt.title('Survival Based on Embarkation Point')
plt.xlabel('Embarkation Point')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

# Survival Based on Fare
plt.figure(figsize=(10, 6))
sns.histplot(data_filled, x='Fare', hue='Survived', bins=20, kde=True, palette='viridis')
plt.title('Fare Distribution by Survival')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()

# Family Size Analysis
data_filled['Family_Size'] = data_filled['SibSp'] + data_filled['Parch']

plt.figure(figsize=(10, 6))
sns.countplot(x='Family_Size', hue='Survived', data=data_filled, palette='cubehelix')
plt.title('Survival Based on Family Size')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

# Feature Engineering

```python
# Create Age Group
bins = [0, 12, 18, 40, 60, 80]
labels = ['Child', 'Teenager', 'Adult', 'Middle-aged', 'Senior']
data_filled['Age_Group'] = pd.cut(data_filled['Age'], bins=bins, labels=labels)

# Visualize Survival by Age Group
plt.figure(figsize=(10, 6))
sns.countplot(x='Age_Group', hue='Survived', data=data_filled, palette='coolwarm')
plt.title('Survival Based on Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

# Titanic Dataset EDA Conclusion

After exploring the Titanic dataset, key insights include:

- **Gender:** Females showed a much higher survival rate than males, reflecting evacuation priorities.

- **Age:** Children had the best survival chances; survival declined with age.

- **Passenger Class:** First-class passengers had notably higher survival rates than second and third class.

- **Fare:** Higher fares correlated with increased survival, indicating economic advantage.

- **Family Size:** Small families had better survival odds, while larger families faced more risk.

- **Embarkation Point:** Passengers boarding at Cherbourg had higher survival compared to Southampton and Queenstown.

- **Correlations:** Survival correlated positively with fare and passenger class.

Handling missing values with imputation improved data quality, supporting these findings.

These insights highlight how social factors and demographics influenced survival on the Titanic.
